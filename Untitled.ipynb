```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import os
# set random seed
SEED = 321
np.random.seed(SEED)
```

# Part 1

```python
df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vQN3Ksa9szQuO4G1-msXWAp17KtVHQCBnuEieu_auu1wSiBf3-krHusIx5VBMkihxj-KZLBosDIGEyR/pub?output=csv')
df
```

```python
# Filter the 4 largest cities
df_top = df[df['SizeRank'] < 4]
```

```python
# Pulling ID columns
id_vars = df_top.columns[:8]
id_vars
```

```python
# Melt the home data 
long_home = pd.melt(df_top,id_vars=id_vars)
long_home
```

```python
## Set the var_name  as "Date"
long_home = pd.melt(df_top,id_vars=id_vars, var_name='Date', 
                      value_name="Value")
long_home
```

```python
# Convert the dates into datetime datatype
long_home['Date'] = pd.to_datetime(long_home['Date'])
```

```python
# Set the datetime column as the index
long_home.set_index('Date', inplace=True)
```

```python
long_home.index
```

```python
# Resample the dataframe as monthly frequency, grouped by City
long_home_resampled = long_home.groupby('RegionName').resample('M').mean()
```

# Part 2

```python
# Create a pivot of the home values
long_home_pivot = long_home_resampled.pivot_table(index='Date', columns='RegionName', values='Value')
long_home_pivot
```

```python
from matplotlib.ticker import FuncFormatter

def thousands(x,pos):
    """function for use wth matplotlib FuncFormatter -  formats money in millions"""
    new_x = x/1_000
    return f"{new_x:,.0f}K"
```

```python
# Plotting our newly created pivot table
ax = long_home_pivot.plot()

# Add a title and axis labels
ax.set_title('Zillow Home Value Index for 4 Largest Cities')
ax.set_xlabel('Date')
ax.set_ylabel('Home Value')

ax.yaxis.set_major_formatter(FuncFormatter(thousands))

plt.show();
```
